```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
dataframe = pd.read_csv(r"dataset_cajibio_promediado.csv")
dataframe.head()
```

```python
abril = "-04-"
mayo = "-05-"
junio = "-06-"
posicion = -1
NumEtapa = -1     #Para el año 2005, el número de etapa sería el 0. Para 2020, NumEtapa=15
año = [None]*16
tiempoTermico = [0]*16
```

```python
for fecha in dataframe['Fecha']:
    posicion = posicion + 1
    if abril in fecha:
        if abril in dataframe.loc[posicion-1,'Fecha']:
            newTempMedia = dataframe.loc[posicion,'temp_media'] - 10
            tiempoTermico[NumEtapa] = tiempoTermico[NumEtapa] + newTempMedia
            
        else:  # Este else sólo va a ocurrir los días primero de Abril.
            NumEtapa = NumEtapa + 1
            año[NumEtapa]=fecha[0:4]
    if mayo in fecha:
        newTempMedia = dataframe.loc[posicion,'temp_media'] - 10
        tiempoTermico[NumEtapa] = tiempoTermico[NumEtapa] + newTempMedia
    if junio in fecha:
        newTempMedia = dataframe.loc[posicion,'temp_media'] - 10
        tiempoTermico[NumEtapa] = tiempoTermico[NumEtapa] + newTempMedia
```

```python
tiempoTermicoDF = DataFrame(tiempoTermico, columns=['tiempo_termico'])
añoDF = DataFrame(año, columns=['año'])
```

```python
df = pd.concat([añoDF, tiempoTermicoDF],axis=1)
df.head(16)
```
