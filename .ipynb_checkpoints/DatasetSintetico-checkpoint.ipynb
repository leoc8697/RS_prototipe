```python
    import csv
    import pandas as pd
    import numpy as np
    n_samples = 100
```

```python
def genLluvia(n_samples, devEstandar, media):

    lluvia = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((lluvia)))
    return res
```

```python
def genHumedadMediaAire(n_samples, devEstandar, media):

    humedadMediaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMediaAire)))
    return res

def genHumedadMinimaAire(n_samples, devEstandar, media):

    humedadMinimaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMinimaAire)))
    return res

def genHumedadMaximaAire(n_samples, devEstandar, media):

    humedadMaximaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMaximaAire)))
    return res

def genHumedadMediaSuelo(n_samples, devEstandar, media):

    humedadMediaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMediaSuelo)))
    return res

def genHumedadMinimaSuelo(n_samples, devEstandar, media):

    humedadMinimaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMinimaSuelo)))
    return res

def genHumedadMaximaSuelo(n_samples, devEstandar, media):

    humedadMaximaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((humedadMaximaSuelo)))
    return res

def genPh(n_samples, devEstandar, media):

    ph = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((ph)))
    return res

def genPromHoraSol(n_samples, devEstandar, media):

    promHoraSol = devEstandar * np.random.randn(1,n_samples) + media
    print(promHoraSol)
    res = pd.DataFrame(data=np.column_stack((promHoraSol)))
    return res

def genNitrogenoSuelo(n_samples):

    nitrogenoSuelo = 0
    for i in range(n_samples):
         nitrogenoSuelo = np.append(nitrogenoSuelo, np.random.choice((100,200,300,400,500)))
    nitrogenoSuelo = np.delete(nitrogenoSuelo, 0, axis=0)
    #nitrogenoSuelo = nitrogenoSuelo.reshape(100,1)
    print(nitrogenoSuelo)
    res = pd.DataFrame(data=np.row_stack((nitrogenoSuelo)))
    #res = res.reshape(100,1)
    return res

def genTempMediaAire(n_samples, devEstandar, media):

    tempMediaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMediaAire)))
    return res

def genTempMinimaAire(n_samples, devEstandar, media):

    tempMinimaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMinimaAire)))
    return res

def genTempMaximaAire(n_samples, devEstandar, media):

    tempMaximaAire = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMaximaAire)))
    return res

def genTempMinimaSuelo(n_samples, devEstandar, media):

    tempMinimaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMinimaSuelo)))
    return res

def genTempMaximaSuelo(n_samples, devEstandar, media):

    tempMaximaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMaximaSuelo)))
    return res

def genTempMediaSuelo(n_samples, devEstandar, media):

    tempMediaSuelo = devEstandar * np.random.randn(1,n_samples) + media
    res = pd.DataFrame(data=np.column_stack((tempMediaSuelo)))
    return res
```

```python
for i in range(n_samples):
     day = np.array([range(n_samples)])
     day = np.append(day, i+1)
     day = np.delete(day, 0, axis=0)
```

```python
dataLluvia = genLluvia(n_samples, 20, 30)
dataHumedadMediaAire = genHumedadMediaAire(n_samples, 20, 65)
dataHumedadMinimaAire = genHumedadMinimaAire(n_samples, 5, 40)
dataHumedadMaximaAire = genHumedadMaximaAire(n_samples, 5, 80)
dataHumedadMediaSuelo = genHumedadMediaSuelo(n_samples, 20, 65)
dataHumedadMinimaSuelo = genHumedadMinimaSuelo(n_samples, 5, 40)
dataHumedadMaximaSuelo = genHumedadMaximaSuelo(n_samples, 5, 80)
dataPh = genPh(n_samples, 1.5, 6)
dataPromHoraSol = genPromHoraSol(n_samples, 3, 7)
dataNitrogenoSuelo = genNitrogenoSuelo(n_samples)
dataTempMediaAire = genTempMediaAire(n_samples, 5, 21)
dataTempMinimaAire = genTempMinimaAire(n_samples, 2, 16)
dataTempMaximaAire = genTempMaximaAire(n_samples, 2, 28)
dataTempMediaSuelo = genTempMediaSuelo(n_samples, 5, 18)
dataTempMinimaSuelo = genTempMinimaSuelo(n_samples, 2, 14)
dataTempMaximaSuelo = genTempMaximaSuelo(n_samples, 2, 26)
```

```python
def genDataset(res):

    #p = np.random.randint(low = low, high = high, size=(1, n_samples))
    res = pd.DataFrame(data=np.column_stack((day,dataLluvia, dataHumedadMediaAire, dataHumedadMinimaAire,
                                             dataHumedadMaximaAire, dataHumedadMediaSuelo, dataHumedadMinimaSuelo,
                                             dataHumedadMaximaSuelo, dataPh,dataPromHoraSol, dataNitrogenoSuelo,
                                             dataTempMediaAire, dataTempMinimaAire, dataTempMaximaAire,
                                             dataTempMediaSuelo, dataTempMinimaSuelo, dataTempMaximaSuelo)),
            columns = ['Dia','LluviaMediaDiaria(cm^2/cm^3)', 'HumedadMediaAire(%)', 'HumedadMinimaAire(%)',
                      'HumedadMaximaAire(%)', 'HumedadMediaSuelo(%)', 'HumedadMinimaSuelo(%)',
                      'HumedadMaximaSuelo(%)','pH', 'PromedioHorasSol', 'NitrogenoSuelo(mg/Kg)',
                      'TemperaturaMediaAire(°C)', 'TemperaturaMinimaAire(°C)', 'TemperaturaMaximaAire(°C)',
                      'TemperaturaMediaSuelo(°C)', 'TemperaturaMinimaSuelo(°C)', 'TemperaturaMaximaSuelo(°C)'])
    return res
```
